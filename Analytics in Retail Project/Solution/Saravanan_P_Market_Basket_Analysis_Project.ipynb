Perform grouping in Grocery Dataset

```python
data_unstack= data.groupby(['Member_number','itemDescription'])['itemDescription'].count().unstack().fillna(0)
data_unstack
```

Encode the data"s in Dataset

```python
def encode(x):
  if x>0:
    return 1
  if x<=0:
    return 0
data_code=data_unstack.applymap(encode)
```

```python
data_code
```

Apply Apriori Algorithm

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
apr= apriori(data_code,min_support=0.07,use_colnames=True)
apr
```

Apply Association Sales

```python
data_rules = association_rules(apr,metric='support',min_threshold=0.07)
data_rules
```

```python
def get_recommendations(product,rules,num_recommendation=5):
  recommendations = []
  for idx,row in rules.iterrows():
    if product in row['antecedents']:
      recommended_item = next(iter(row['consequents']))
      recommendations.append((recommended_item,row['lift']))
  recommendations.sort(key=lambda x: x[1],reverse=True)
  return recommendations[0:num_recommendation]
```

```python
product_to_recommend='tropical fruit'
recommendations=get_recommendations(product_to_recommend,data_rules)
print(f"recommendations for'{product_to_recommend}':")
for item,lift in recommendations:
  print(f"{item} (lift: {lift:.2f})")
```
